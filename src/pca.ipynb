```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
X_train = pd.read_csv('../../data/train_in.csv', header=None)
y_train = pd.read_csv('../../data/train_out.csv', header=None).values.flatten()
```

```python
print(X_train)
print(y_train)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_train)
```

```python
print(pca.explained_variance_ratio_)
print(X_pca)
```

```python
plt.figure(figsize=(20, 10))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_train, cmap='tab10', alpha=0.3)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('PCA of MNIST digits')
plt.colorbar(scatter, ticks=range(10), label='Digits')
plt.show()
```
