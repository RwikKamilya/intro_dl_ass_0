most difficult to classify correctly? Again, for calculating and visualising confusion matrices you may
use the sklearn package. Describe your findings, and compare the performance of your classifiers on
the train and test sets.

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

import sklearn
from sklearn.neighbors import NearestCentroid
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Read data
train_in = pd.read_csv("../../data/train_in.csv").values
test_in = pd.read_csv("../../data/test_in.csv").values
train_out = pd.read_csv("../../data/train_out.csv").values.ravel()
test_out = pd.read_csv("../../data/test_out.csv").values.ravel()
```

```python
# Nearest Mean Classifier

# Train and test classiffier:
NearestMeanCLF = NearestCentroid()
NearestMeanCLF.fit(train_in, train_out)
pred1 = NearestMeanCLF.predict(test_in)

# Print accuracy
print(f"Accuracy: {sklearn.metrics.accuracy_score(test_out, pred1):.3f}")

# Display confusion matrix
sklearn.metrics.ConfusionMatrixDisplay.from_predictions(
    test_out,
    pred1,
    normalize='true',
    values_format='.2f'
)
plt.title("Confusion Matrix of Nearest Mean Classifier")
plt.show()
```

```python
# KNN Classifier

# Train and test classiffier:
KnnCLF = KNeighborsClassifier(n_neighbors=1)
KnnCLF.fit(train_in, train_out)
pred2 = KnnCLF.predict(test_in)

# Print accuracy
print(f"Accuracy: {sklearn.metrics.accuracy_score(test_out, pred2):.3f}")

# Display confusion matrix
sklearn.metrics.ConfusionMatrixDisplay.from_predictions(
    test_out,
    pred2,
    normalize='true',
    values_format='.2f'
)
plt.title("Confusion Matrix of KNN Classifier")
plt.show()
```
